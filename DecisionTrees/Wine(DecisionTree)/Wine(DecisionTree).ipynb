```python
from sklearn import tree
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
#Get example wine data
wine = load_wine()
```

```python
wine.data
```

```python
# Identify how many classifers 
wine.target
```

```python
# Combine Datasets
pd.concat([pd.DataFrame(wine.data),pd.DataFrame(wine.target)],axis = 1)
```

```python
wine_featurenames= wine.feature_names
```

```python
wine_targetnames=wine.target_names
```

```python
#Train model
Xtrain,Xtest,Ytrain,Ytest = train_test_split(wine.data,wine.target,test_size = 0.3)
```

```python
#Create instance for the tree
clf = tree.DecisionTreeClassifier(criterion = "entropy")
clf = clf.fit(Xtrain,Ytrain)
score = clf.score(Xtest,Ytest) 
```

```python
#It's a good score, means it fitts well with the data
score
```

```python
#Garph
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 10))
tree.plot_tree(
    clf,
    feature_names=wine.feature_names,   
    class_names=wine.target_names,      
    filled=True,
    rounded=True
)
plt.tight_layout()
plt.show()
```

```python
# It's now showing clearly
clf.feature_importances_
```

```python
# By doing this,we can get to know how important each feature is at a glance
[*zip(wine_featurenames,clf.feature_importances_)]
```
