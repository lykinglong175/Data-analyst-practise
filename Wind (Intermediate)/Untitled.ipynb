```python
import pandas as pd
import datetime
data_url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data'
wind=pd.read_csv(data_url,sep=r'\s+',parse_dates=[[0,1,2]])
wind.head()
```

```python
#Fixing year 2061
def fix_year(x):
    year= x.year - 100 if x.year > 1989 else x.year
    return datetime.date(year,x.month,x.day)
wind['Yr_Mo_Dy'] = wind['Yr_Mo_Dy'].apply(fix_year)
wind.head()
```

```python
#Set the right dates as the index
wind['Yr_Mo_Dy'] = pd.to_datetime(wind['Yr_Mo_Dy'])
wind = wind.set_index('Yr_Mo_Dy')
wind.head()
```

```python
#Compute how many values are missing for each location over the entire record.
wind.isnull().sum()
```

```python
#Compute how many non-missing values there are in total
wind.notnull().sum()
```

```python
#Calculate the mean windspeeds of the windspeeds over all the locations and all the times.
wind.sum().sum()/wind.notna().sum().sum().astype(int)
```

```python
# Create a DataFrame called loc_stats and calculate 
# the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days
loc_stats = wind.describe(percentiles=[])
loc_stats
```

```python
#Create a DataFrame called day_stats and calculate 
#the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day.
day_stats = pd.DataFrame()
day_stats['min'] = wind.min(axis=1)
day_stats ['max']= wind.max(axis=1)
day_stats ['mean']= wind.mean(axis=1)
day_stats ['std']= wind.std(axis=1)
day_stats.head()
```

```python
#Find the average windspeed in January for each location.
wind.loc[wind.index.month == 1].mean()
```

```python
#Downsample the record to a yearly frequency for each location
wind.groupby(wind.index.to_period('A')).mean()
```

```python
#Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations
#for each week (assume that the first week starts on January 2 1961) for the first 52 weeks.
weekly = wind.resample('W').agg(['min','max','mean','std'])
weekly.loc[weekly.index[1:53],'RAT':'MAL'].head()
```
