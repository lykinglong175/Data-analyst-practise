```python
import pandas as pd 
import matplotlib.pyplot as plt
from collections import Counter

%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(url, sep = '\t')
chipo.head()
```

```python
# A histogram of the top 5 items bought
names = chipo.item_name
counts = names.value_counts()
counts = counts.sort_values()[45:50]
counts.plot(kind='bar',title='TOp 5 Item Bought')
plt.xlabel('item')
plt.ylabel('Count')
plt.show()
```

```python
# A scatterplot with the number of items orderered per order price
chipo.item_price = [float(str(num).replace('$','').strip()) for num in chipo.item_price]
orders = chipo.groupby('order_id').sum()
plt.scatter(x= orders.item_price,y=orders.quantity,s=50,c='blue')
plt.xlabel('Order Price')
plt.ylabel('Item Ordered')
plt.title('Number of items ordered per order price')
```
