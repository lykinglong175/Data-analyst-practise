```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python


titan = pd.read_csv('train.csv')

titan.head()
```

```python
titan.set_index('PassengerId').head()
```

```python
males = (titan['Sex'] =='male').sum()
females = (titan['Sex'] =='female').sum()
proportion=[males,females]
```

```python
plt.pie(proportion,labels = ['Male','Female'],colors=['blue','red'])
plt.title("Sex Proportion")
plt.show()
```

```python
lm = sns.lmplot(x='Age',y='Fare',data=titan,hue='Sex')
lm.set(title='Fare x Age')

axes= lm.axes
axes[0,0].set_ylim(-5,)
axes[0,0].set_xlim(-5,85)
```
